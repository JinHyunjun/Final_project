```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_2023 = pd.read_csv("C:\\multi_final\\data\\2023_combine.csv")
df_2023
```

```python
# 대여소 번호별 총 거치대 수를 사전으로 정의
total_docks = {
    1124: 6,
    1153: 17,
    1158: 20,
    1160: 20,
    1166: 10,
    2701: 26,
    2715: 25,
    2721: 12,
    2728: 55,
    3798: 11
}
```

```python
# 거치대 수량을 해당 대여소의 총 거치대 수로 나누는 함수 정의
def calculate_dock_ratio(row):
    station_id = row['대여소번호']
    if station_id in total_docks:
        total_dock = total_docks[station_id]
        return row['거치대수량'] / total_dock
    else:
        return None  # 대여소 번호가 사전에 없는 경우 None 반환
```

```python
df_2023['거치대비율'] = df_2023.apply(calculate_dock_ratio, axis=1)
```

```python
# '일시'와 '시간대'를 결합하여 'datetime' 열 생성
df_2023['datetime'] = pd.to_datetime(df_2023['일시']) + pd.to_timedelta(df_2023['시간대'], unit='h')
```

```python
# '대여소번호'를 'ID'로, '거치대수량'을 'count'로 이름 변경
df_2023.rename(columns={'대여소번호': 'ID', '거치대수량': 'count', '일시' : 'date', '시간대' : 'hour', '거치대비율':'ratio'}, inplace=True)
```

```python
# 필요한 열만 선택
df_2023 = df_2023[['ID', 'date', 'hour', 'count', 'datetime', 'ratio']]
df_2023
```

```python
from sklearn.preprocessing import MinMaxScaler

# 정규화를 위한 스케일러 초기화
scaler = MinMaxScaler()

df_2023['ratio_scaled'] = scaler.fit_transform(df_2023['ratio'].values.reshape(-1,1))
# 결과 확인
print(df_2023.head())
```

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
import numpy as np

```

```python
# 시퀀스 데이터로 변환하는 함수
def create_dataset(X, y, look_back=1):
    Xs, ys = [], []
    for i in range(len(X) - look_back):
        v = X.iloc[i:(i + look_back)].values
        Xs.append(v)
        ys.append(y.iloc[i + look_back])
    return np.array(Xs), np.array(ys)

look_back = 3  # 과거 3개의 시간 단계를 사용하여 다음 단계를 예측
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
```

```python
# 저장된 LSTM 모델 불러오기
model = load_model('C:\\multi_final\\deep_learning_code\\my_model.keras')
```

```python
# ID 1124에 대한 데이터만 추출
stop_data = df_2023[df_2023['ID'] == 1124]

# 입력 변수와 예측할 변수
X = stop_data[['datetime', 'ratio_scaled']]
y = stop_data['ratio_scaled']
```

```python
# 시퀀스 데이터 생성
X_seq, y_seq = create_dataset(X, y, look_back)

# 저장된 LSTM 모델 불러오기
model = load_model('C:\\multi_final\\deep_learning_code\\my_model.keras')
```

```python
# 모델을 이용해 예측하기
predictions = model.predict(X_seq)

print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# 실제 값
actual = y_seq

# MSE 계산
mse = mean_squared_error(actual, predictions)
print(f'MSE: {mse}')

# RMSE 계산
rmse = sqrt(mse)
print(f'RMSE: {rmse}')
```

```python
from sklearn.metrics import r2_score

# R squared 계산
r2 = r2_score(actual, predictions)
print(f'R²: {r2}')
```
