```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\multi_final\\data\\2022년.csv", encoding='utf-8')
df
```

```python
# '시간대' 칼럼을 정수형으로 변환
df['시간대'] = df['시간대'].astype(int)

# 100이 넘는 값들에 대해 100으로 나눈 후 몫을 해당 값으로 대체
df['시간대'] = df['시간대'].apply(lambda x: x // 100 if x > 100 else x)
```

```python
# 결과 확인
print(df[['일시', '대여소번호', '대여소명', '시간대', '거치대수량', 'addr1', 'addr2', '기온', '강수형태', '강수량', '총생활인구수']])
```

```python
# 데이터 프레임을 새로운 CSV 파일로 저장
df.to_csv("C:\\multi_final\\data\\2022년_수정.csv", encoding='utf-8', index=False)
```
